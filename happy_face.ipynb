# Happy Face Plot
This notebook uses matplotlib to draw a simple happy face.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Face outline
face = Circle((0, 0), 1, facecolor='gold', edgecolor='black', linewidth=2)
ax.add_patch(face)

# Eyes
left_eye = Circle((-0.35, 0.3), 0.1, color='black')
right_eye = Circle((0.35, 0.3), 0.1, color='black')
ax.add_patch(left_eye)
ax.add_patch(right_eye)

# Smile
smile = Arc((0, -0.05), 1.1, 0.8, angle=0, theta1=200, theta2=340, linewidth=3, color='black')
ax.add_patch(smile)

ax.set_aspect('equal')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.axis('off')
plt.show()
```
